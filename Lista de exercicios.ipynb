#### EX 1 
Crie uma classe Aluno que armazene o nome e uma lista de notas. Implemente um 
método para calcular a média e outro para verificar se o aluno foi aprovado (média>= 6).

```python
class Aluno:
    def __init__(self, nome, nota):
        self.nome = nome
        self.nota = nota

    def saudar(self):
        return f"Olá, meu nome é {self.nome};"

p1 = Aluno("Vitor", [7.5, 8.0, 6.5])

soma = sum(p1.nota)
media = soma/len(p1.nota)
if media >= 6:
    resultado = "Aprovado"
else:
    resultado = "Exame"
print(f"{p1.saudar()} Minha média é {media:.2f} e estou {resultado}.")
```

#### Ex2

Crie uma classe Conversor que possua um atributo privado taxa_dolar. Implemente
métodos para atualizar a taxa (com validação para não aceitar valores negativos) e
para converter um valor de Real para Dólar.

```python
class Conversor:
    
    def __init__(self, taxa_dolar):
        # validação da taxa inicial
        if taxa_dolar <= 0:
            raise ValueError("A taxa deve ser maior que zero.")
        
        # atributo privado
        self.__taxa_dolar = taxa_dolar

    def atualizar_taxa(self, nova_taxa):
        if nova_taxa <= 0:
            print("Erro: a taxa não pode ser negativa ou zero.")
        else:
            self.__taxa_dolar = nova_taxa
            print("Taxa atualizada com sucesso!")

    def converter_para_dolar(self, valor_real):
        return valor_real / self.__taxa_dolar


conversor = Conversor(5.0)

print("Valor convertido:", conversor.converter_para_dolar(100))

conversor.atualizar_taxa(5.2)

print("Valor convertido com nova taxa:", f"{conversor.converter_para_dolar(100):.2f}")
    

    
```

#### Ex3

Crie as classes Livro e Biblioteca. A classe Biblioteca deve conter uma lista de
objetos do tipo Livro. Implemente métodos para adicionar livros e listar todos os
títulos disponíveis.

```python
class Livro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor


class Biblioteca:
    def __init__(self):
        self.livros = []  # lista para guardar objetos Livro

    def adicionar_livro(self, livro):
        self.livros.append(livro)
        print(f'Livro "{livro.titulo}" adicionado com sucesso!')

    def listar_livros(self):
        print("Livros disponíveis:")
        for livro in self.livros:
            print(f'- {livro.titulo}')


# Testando
livro1 = Livro("Dom Casmurro", "Machado de Assis")
livro2 = Livro("1984", "George Orwell")

biblioteca = Biblioteca()
biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_livro(livro2)

biblioteca.listar_livros()
```

Crie uma classe base Veiculo com atributos como marca e modelo. Crie
subclasses Carro (com número de portas) e Moto (com cilindradas). Ambas devem
herdar as características da classe pai.

```python
class Veiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo


class Carro(Veiculo):
    def __init__(self, marca, modelo, portas):
        super().__init__(marca, modelo)
        self.portas = portas


class Moto(Veiculo):
    def __init__(self, marca, modelo, cilindradas):
        super().__init__(marca, modelo)
        self.cilindradas = cilindradas


# Testando
carro = Carro("Toyota", "Corolla", 4)
moto = Moto("Honda", "CB500", 500)

print(carro.marca, carro.modelo, carro.portas)
print(moto.marca, moto.modelo, moto.cilindradas)
```

No contexto de IoT, crie uma classe Sensor que armazene ID e tipo. Crie um script
que permita ao usuário cadastrar vários sensores em um dicionário, onde a chave
é o ID e o valor é o objeto Sensor.

```python
class Sensor:
    def __init__(self, id_sensor, tipo):
        self.id = id_sensor
        self.tipo = tipo


sensores = {}

# Cadastro simples
for i in range(2):  # pode alterar quantidade
    id_sensor = input("Digite o ID do sensor: ")
    tipo = input("Digite o tipo do sensor: ")
    
    sensor = Sensor(id_sensor, tipo)
    sensores[id_sensor] = sensor

print("\nSensores cadastrados:")
for chave, sensor in sensores.items():
    print(f'ID: {chave} | Tipo: {sensor.tipo}')
```

#### Ex9

Crie uma classe Funcionario que possua um método registrar_entrada(). Utilize o
módulo datetime para capturar o horário exato em que o método foi chamado e
armazene isso em uma lista de logs dentro do objeto.

```python
from datetime import datetime

class Funcionario:
    def __init__(self, nome):
        self.nome = nome
        self.logs = []

    def registrar_entrada(self):
        horario = datetime.now()
        self.logs.append(horario)
        print(f"Entrada registrada às {horario}")


# Testando
func = Funcionario("Vitor")
func.registrar_entrada()
func.registrar_entrada()

print("Logs:", func.logs)
```

#### Ex10

Crie uma classe Calculadora que contenha apenas métodos estáticos para
operações como soma, subtração e raiz quadrada. O objetivo é usar a classe como
um "namespace" de funções utilitárias sem precisar instanciá-la.



```python
import math

class Calculadora:

    @staticmethod
    def soma(a, b):
        return a + b

    @staticmethod
    def subtracao(a, b):
        return a - b

    @staticmethod
    def raiz_quadrada(x):
        return math.sqrt(x)


# Testando (sem instanciar a classe)
print(Calculadora.soma(10, 5))
print(Calculadora.subtracao(10, 5))
print(Calculadora.raiz_quadrada(25))
```
